```python
import numpy as np
from scipy.optimize import fsolve
from math import pi
import matplotlib.pyplot as plt
```

```python
X = np.arange(0,360,0.01)
Y = 12*(X % 30)

plt.plot(X,Y)
plt.plot(X,X+180)
plt.plot(X,X-180)
plt.show()
```

```python
def func(z):
    return [12*(z[0]%30)-z[1], abs(z[0]-z[1])-180 ]
```

```python
time = fsolve(func, [60,60])
```

```python
print('{} hours {} minutes {} seconds'.format(int(time[0]/30), int(time[1]/6), int((time[1]%6)*10)))
```

```python
for i in range(11):
    time = fsolve(func, [i*30,i*30])
    print('{} hours {} minutes {} seconds'.format(int(time[0]/30), int(time[1]/6), int((time[1]%6)*10)))
```
